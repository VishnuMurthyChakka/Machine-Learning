```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
```

```python
# create clusters of data
from sklearn.datasets.samples_generator import make_blobs
X, Y = make_blobs(n_samples=[950, 50], centers=None, n_features=2,
                   random_state=0)
print(X.shape)
print(Y.shape)
```

```python
Counter(Y)
```

```python
X = pd.DataFrame(X,columns = ["x1", "x2"])
print(X.shape)
```

```python
plt.scatter(X.x1,X.x2,
           c = Y,
           cmap = 'viridis')
plt.show()
```

# Under Sampling

```python
from imblearn.under_sampling import RandomUnderSampler
RUS = RandomUnderSampler(random_state=1)
X_Under, Y_Under = RUS.fit_sample(X, Y)
print(X_Under.shape)
print(Y_Under.shape)
print(Counter(Y_Under))
```

```python
Y_Under
```

```python
plt.scatter(X_Under.x1,X_Under.x2,
           c = Y_Under,
           cmap = 'viridis')
plt.show()
```

# Over Sampling

```python
# Over sampling
from imblearn.over_sampling import RandomOverSampler
ROS = RandomOverSampler(random_state=0)
X_Over, Y_Over = ROS.fit_sample(X, Y)
print(X_Over.shape)
print(Y_Over.shape)
print(Counter(Y_Over))
```

```python
plt.scatter(X_Over.x1,X_Over.x2,
           c = Y_Over,
           alpha = 0.7, 
           cmap = 'viridis')
plt.show()
```
