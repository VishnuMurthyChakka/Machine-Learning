```python
import matplotlib.pyplot as plt  
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import make_circles
X,Y = make_circles(n_samples=100,factor=0.5)
X = pd.DataFrame(X,columns = ["x1", "x2"])
print(X.shape)
print(X.columns)
```

```python
np.unique(Y)
```

```python
plt.scatter(X.x1,X.x2,
           c = Y,
           cmap = 'viridis')
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
DTC = DecisionTreeClassifier(max_depth=2)
DTC.fit(X,Y)
```

```python
predict = DTC.predict(X)
print(predict)
```

```python
plt.scatter(X.x1,X.x2,
           c = predict,
           cmap = 'viridis')
plt.show()
```

```python
from sklearn import tree
tree.plot_tree(decision_tree = DTC,
               feature_names = X.columns.values,
               class_names = True,
               filled=True,
               node_ids=True,
               proportion=False)
```
